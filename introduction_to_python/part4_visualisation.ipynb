# Visualisation

The main package to know is matplotlib. It is the biggest, most popular plotting pacakge. Even some other popular tools like seaborn are based on matplotlib

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# plt.plot()

This function does you line plot, and your scatter plot

```python
plt.plot([1,2,3],[2,4,5])
```

## plotting multiple lines

```python
x = np.linspace(0,7,100) # this creates an array from 0 7 in 100 steps.
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

# Add a legend

```python
x = np.linspace(0,7,100)
plt.plot(x,np.sin(x),label="sin()")
plt.plot(x,np.cos(x),label='cos()')
plt.legend()
```

# title, xlabel, ylabel

```python
x = np.linspace(0,7,100)
plt.plot(x,np.sin(x),label="sin()")
plt.plot(x,np.cos(x),label='cos()')
plt.legend()
plt.title("TITLE text")
plt.xlabel("xlabel text")
plt.ylabel("ylabel text")
```

# subplots and different linestyles

```python
x = [1,2,3]
y = [2,4,5]

fig, ax = plt.subplots(2, 2) # 2 rows, 2 columns

ax[0,0].plot(x,y,'o')
ax[0,1].plot(x,y,'r--')
ax[1,0].plot(x,y,'y*--')
ax[1,1].plot(x,y,'c')
```

you can see there are these shorthand descriptions of the type of line plot you want. `o` gives scatter plot, `y*--` gives yellow line, star points, and dashed line

# Pandas plotting

You can plot stuff directly from pandas too

```python
# lets visualise the titanic dataset again
df = pd.read_csv("titanic_dataset.csv")
```

```python
df["age"].hist()
plt.xlabel("Age")
```

```python
group_by_sex =df.groupby("sex")["alive"].count()
group_by_sex
```

```python
group_by_sex.plot.bar()
```

# Seaborn

```python
# df.pivot_table(columns="sex",index="age",values="su")
ax = sns.countplot(x = 'embarked', hue = 'survived',data = df)
ax.set(title = 'Survival distribution according to Embarking place')
plt.show()
# s: southampton, c:cherbourg, q:quenstown
```

```python
sns.violinplot(x="sex", y="age",hue="survived", split=True, data=df)
# here we are looking at the distribution of ages for males, females. 
# we are comparing dead (blue) with survived(orange) distributions
```
